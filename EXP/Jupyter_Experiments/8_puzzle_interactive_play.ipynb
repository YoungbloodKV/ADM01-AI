```python
# Option 2: Interactive Play (Manual Solver)

goal_state = [[1, 2, 3],
              [4, 5, 6],
              [7, 8, 0]]  # 0 = empty tile (shown as _)

def print_state(state):
    for row in state:
        print(" ".join(str(x) if x != 0 else "_" for x in row))
    print()

def find_zero(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

def interactive_play(start_state):
    state = [row[:] for row in start_state]  # copy
    print("Interactive Play: Use W/A/S/D to move the empty tile (shown as _).")
    print("Goal State:")
    print_state(goal_state)

    while state != goal_state:
        print_state(state)
        move = input("Move (W=up, S=down, A=left, D=right, Q=quit): ").lower()
        
        if move == "q":
            print("Game quit.")
            return
        
        x, y = find_zero(state)
        if move == "w" and x > 0:   # move blank up
            state[x][y], state[x - 1][y] = state[x - 1][y], state[x][y]
        elif move == "s" and x < 2: # move blank down
            state[x][y], state[x + 1][y] = state[x + 1][y], state[x][y]
        elif move == "a" and y > 0: # move blank left
            state[x][y], state[x][y - 1] = state[x][y - 1], state[x][y]
        elif move == "d" and y < 2: # move blank right
            state[x][y], state[x][y + 1] = state[x][y + 1], state[x][y]
        else:
            print("Invalid move!")
    
    print("ğŸ‰ Congratulations! You solved it!")
    print_state(state)


# Example Start State
start_state = [[1, 2, 3],
               [4, 0, 6],
               [7, 5, 8]]

# Run interactive mode
interactive_play(start_state)
```
