```python
# Experiment 3: Water Jug Problem (BFS)

from collections import deque

def print_jugs(x, y, jug1, jug2):
    """Pretty print jug states with icons"""
    jug_icon = "ðŸ«™"  # jug emoji
    print(f"Jug1: {x}/{jug1} {jug_icon * x}")
    print(f"Jug2: {y}/{jug2} {jug_icon * y}")
    print("-" * 30)

def water_jug_solver(jug1, jug2, target):
    visited = set()
    q = deque()
    
    # Start with both jugs empty
    q.append((0, 0))
    
    print("============================================")
    print(" Water Jug Problem Execution (BFS Approach)")
    print(" Executed by: KEERTHIVASAN.S, Register No: 192553013")
    print("============================================\n")
    
    while q:
        x, y = q.popleft()
        
        if (x, y) in visited:
            continue
        visited.add((x, y))
        
        # Show jug states with icons
        print_jugs(x, y, jug1, jug2)
        
        # Check target condition
        if x == target or y == target:
            print("ðŸŽ¯ Target achieved!")
            print(f"Final State â†’ Jug1: {x}, Jug2: {y}")
            print("\nProgram executed successfully by KEERTHIVASAN.S (Reg No: 192553013)")
            return
        
        # Possible moves
        moves = [
            (jug1, y),  # Fill jug1
            (x, jug2),  # Fill jug2
            (0, y),     # Empty jug1
            (x, 0),     # Empty jug2
            # Pour jug1 â†’ jug2
            (x - min(x, jug2 - y), y + min(x, jug2 - y)),
            # Pour jug2 â†’ jug1
            (x + min(y, jug1 - x), y - min(y, jug1 - x))
        ]
        
        for move in moves:
            if move not in visited:
                q.append(move)

# Example run: Jug1=4L, Jug2=3L, Target=2L
water_jug_solver(4, 3, 2)
```
